```python
# 유동인구 2019년 4월 ~ 5월
population_1904 = pd.read_csv('/Users/jaeyoungcho/Desktop/2020 금융 빅데이터 챌린지-공모전데이터/데이터/케이티/유동인구일집계_201904.csv',sep = ',',encoding='CP949')
population_1905 = pd.read_csv('/Users/jaeyoungcho/Desktop/2020 금융 빅데이터 챌린지-공모전데이터/데이터/케이티/유동인구일집계_201905.csv',sep = ',',encoding='CP949')
# 유동인구 2020년 4월 ~ 5월
population_2004 = pd.read_csv('/Users/jaeyoungcho/Desktop/2020 금융 빅데이터 챌린지-공모전데이터/데이터/케이티/유동인구일집계_202004.csv',sep = ',',encoding='CP949')
population_2005 = pd.read_csv('/Users/jaeyoungcho/Desktop/2020 금융 빅데이터 챌린지-공모전데이터/데이터/케이티/유동인구일집계_202005.csv',sep = ',',encoding='CP949')
```

```python
# 유동인구 2019년, 2020년
population_19 = pd.concat([population_1904,population_1905],axis = 0).reset_index(drop=True)
population_20 = pd.concat([population_2004,population_2005],axis = 0).reset_index(drop=True)
```

```python
# 유동인구 
temp_19 = copy.copy(population_19)
temp_20 = copy.copy(population_20)
temp_19['일자'] = temp_19['기준일자'].apply(lambda x : x%10000)
temp_20['일자'] = temp_20['기준일자'].apply(lambda x : x%10000)
temp_19.drop(['기준일자','24시간코드','성별코드','연령대구분값','행정동번호'],axis = 1, inplace = True)
temp_20.drop(['기준일자','24시간코드','성별코드','연령대구분값','행정동번호'],axis = 1, inplace = True)

temp_19 = pd.DataFrame(temp_19.groupby('일자').sum()).reset_index()
temp_20 = pd.DataFrame(temp_20.groupby('일자').sum()).reset_index()

temp_All = pd.merge(temp_19,temp_20,on = '일자')
temp_All.columns = ['일자','2019년 유동인구','2020년 유동인구']

date = temp_All['일자'].unique().tolist()
date_dict = {}
for i,v in enumerate(date):
    date_dict[v] = i
temp_All['일자'].replace(date_dict,inplace = True)
```

```python
plt.plot(temp_All['일자'],temp_All['2019년 유동인구'],label = '2019년 유동인구')
plt.plot(temp_All['일자'],temp_All['2020년 유동인구'],label = '2020년 유동인구')
plt.legend(loc = 'best')
plt.show()
```

```python
# 위 데이터들이 이상해서 다른 공모전에서 사용한 데이터를 비교군을 위해 이용('공모전 제출 불가능')
cheat_1904 = pd.read_csv('/Users/jaeyoungcho/Desktop/2020 BigContest - 포스트 코로나 아이디어/데이터 정의서 및     모음/01_유동인구데이터(SK텔레콤)/4개지역_FLOW_AGE_201904.CSV',sep = '|')
cheat_1905 = pd.read_csv('/Users/jaeyoungcho/Desktop/2020 BigContest - 포스트 코로나 아이디어/데이터 정의서 및     모음/01_유동인구데이터(SK텔레콤)/4개지역_FLOW_AGE_201905.CSV',sep = '|')
cheat_2004 = pd.read_csv('/Users/jaeyoungcho/Desktop/2020 BigContest - 포스트 코로나 아이디어/데이터 정의서 및     모음/01_유동인구데이터(SK텔레콤)/4개지역_FLOW_AGE_202004.CSV',sep = '|')
cheat_2005 = pd.read_csv('/Users/jaeyoungcho/Desktop/2020 BigContest - 포스트 코로나 아이디어/데이터 정의서 및     모음/01_유동인구데이터(SK텔레콤)/4개지역_FLOW_AGE_202005.CSV',sep = '|')

cheat_1904 = cheat_1904.groupby('STD_YMD')[cheat_1904.columns[4:]].sum().reset_index()
cheat_1905 = cheat_1905.groupby('STD_YMD')[cheat_1905.columns[4:]].sum().reset_index()

cheat_2004 = cheat_2004.groupby('STD_YMD')[cheat_2004.columns[4:]].sum().reset_index()
cheat_2005 = cheat_2005.groupby('STD_YMD')[cheat_2005.columns[4:]].sum().reset_index()

cheat_1904['total'] = cheat_1904[cheat_1904.columns[1:]].sum(axis=1)
cheat_1905['total'] = cheat_1905[cheat_1905.columns[1:]].sum(axis=1) 
cheat_2004['total'] = cheat_2004[cheat_2004.columns[1:]].sum(axis=1) 
cheat_2005['total'] = cheat_2005[cheat_2005.columns[1:]].sum(axis=1) 

cheat_1904.drop(cheat_1904.columns[1:-1],axis = 1, inplace = True)
cheat_1905.drop(cheat_1905.columns[1:-1],axis = 1, inplace = True)
cheat_2004.drop(cheat_2004.columns[1:-1],axis = 1, inplace = True)
cheat_2005.drop(cheat_2005.columns[1:-1],axis = 1, inplace = True)

cheat_19 = pd.concat((cheat_1904,cheat_1905),axis = 0).reset_index(drop = True)
cheat_20 = pd.concat((cheat_2004,cheat_2005),axis = 0).reset_index(drop = True)

cheat_19['일자'] = cheat_19['STD_YMD'].apply(lambda x : x % 10000)
cheat_20['일자'] = cheat_20['STD_YMD'].apply(lambda x : x % 10000)
cheat_19.drop(['STD_YMD'],axis = 1, inplace = True)
cheat_20.drop(['STD_YMD'],axis = 1, inplace = True)
cheat_19['일자'].replace(date_dict,inplace = True)
cheat_20['일자'].replace(date_dict,inplace = True)
```

```python
plt.plot(cheat_19['일자'],cheat_19['total'],label = '2019')
plt.plot(cheat_20['일자'],cheat_20['total'],label = '2020')
plt.legend(loc = 'best')
plt.show()
```

### 나이 대 별 유동인구

```python
temp_19 = copy.copy(population_19)
temp_20 = copy.copy(population_20)

temp_19['일자'] = temp_19['기준일자'].apply(lambda x : x % 10000)
temp_20['일자'] = temp_20['기준일자'].apply(lambda x : x % 10000)

age_dict = {}
for i,v in enumerate(sorted(temp_19['연령대구분값'].unique().tolist())):
    age_dict[v] = i * 5

temp_19['연령대구분값'].replace(age_dict,inplace = True)
temp_20['연령대구분값'].replace(age_dict,inplace = True)

temp_19.drop(['기준일자','24시간코드','성별코드','행정동번호'],axis = 1,inplace = True)
temp_20.drop(['기준일자','24시간코드','성별코드','행정동번호'],axis = 1,inplace = True)

temp_19 = pd.DataFrame(temp_19.groupby(['일자','연령대구분값']).sum()).reset_index()
temp_20 = pd.DataFrame(temp_20.groupby(['일자','연령대구분값']).sum()).reset_index()

temp_19['일자'].replace(date_dict,inplace = True)
temp_20['일자'].replace(date_dict,inplace = True)
```

```python
age = sorted(temp_19['연령대구분값'].unique().tolist())
a = []
cnt = 0
fig = plt.figure(figsize = (20,80))
for i in range(len(age)):
    a.append(temp_19.loc[temp_19['연령대구분값'] == age[i]].reset_index(drop = True))
    a.append(temp_20.loc[temp_20['연령대구분값'] == age[i]].reset_index(drop = True))
    temp = fig.add_subplot(len(age)+1/2,2,i+1)
    temp.plot(a[cnt]['일자'],a[cnt]['인구수'],label = '2019년')
    cnt += 1
    temp.plot(a[cnt]['일자'],a[cnt]['인구수'],label = '2020년')
    cnt += 1
    temp.set_title('{}세 ~ {}세'.format(age[i],age[i]+4))
    temp.legend(loc = 'best')
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
```

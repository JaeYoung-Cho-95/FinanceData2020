# 매각결과

```python
매각결과 =  pd.read_csv('/Users/jaeyoungcho/Desktop/2020 금융 빅데이터 챌린지-공모전데이터/데이터/한국감정평가사협회/매각결과.csv',sep = ',',encoding = 'CP949')

# 매각결과 : 수집일자, 적재일시, 센터코드 drop, 비고내용에 있는 nan을 공란으로 바꿈 
매각결과.drop(매각결과.columns[-3:],axis = 1,inplace = True)
매각결과['비고내용'].fillna(' ',inplace = True)
```

```python
# 매각내용에서 유찰이면 유찰, 매각가격이면 매각 가격을 뽑아내고 '최종매각가격'으로 정의. 경매계를 구분해서 '경매계' 로 정의 후 매각내용 drop, 경매계, 최종매각가격 추가
temp = 매각결과['매각내용'].tolist()
최종매각가격 = []
경매계 = []
for i in temp:
    k = ''
    for j in range(len(i)-1,-1,-1):
        if i[j] == '찰':
            최종매각가격.append('유찰')
            break
        elif ord(i[j]) <= 100:
            k = k + '{}'.format(i[j])
        else:
            최종매각가격.append(k[::-1])
            break
    경매계.append(i[0:3]) 
```

```python
매각결과['경매 계'] = pd.Series(경매계)
매각결과['최종매각가격'] = pd.Series(최종매각가격)
매각결과.drop(['매각내용'],axis = 1,inplace = True)
```

```python
# 소재지에서 주요 지역만 추출하여 새로운 열 추가
지역 = 매각결과['소재지및내역내용'].tolist()
잉잉 = []
for i in 지역:
    k = ''
    for j in i:
        if ord(j) >= 50:
            k = k + j
        else:
            잉잉.append(k)
            break
매각결과['지역'] = pd.Series(잉잉)
```
